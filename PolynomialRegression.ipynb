# Polynomial Regression

Example with Virtual Shop: web page speed / purchase data:

```python
%matplotlib inline
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

#Simulate page speed  & purchase amount dataset
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

#### Numpy Polyfit Function

nth-degree polynomial model of our data that minimizes squared error

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

#Let's make prediction model as function

def polyfunc(degree):
    #model itself
    p = np.poly1d(np.polyfit(x, y, degree))
    #visualize data
    xp = np.linspace(0, 7, 100)
    plt.scatter(x, y)
    plt.plot(xp, p(xp), c='r')
    plt.title('Polyfit model')
    plt.xlabel('Seconds')
    plt.ylabel('Dollar')
    plt.show()
    
    #appreciate the model with r2
    r2 = r2_score(y, p(x))
    print(f'r2 for degree = {degree}: {r2:.2f}')

polyfunc(4)
```

4 dimention model looks good. r2 = 0.83 is not bad for training model

```python
# Model for degree = 3
polyfunc(3)
```

3 dimention model gives 5% less accurate model

```python
# Model for degree = 5
polyfunc(5)
```

Increasing degree from 4 to 5 gives 3% better model on training set

```python
# Model for degree = 6
polyfunc(6)
```

Increasing degree from 5 to 6 gives 1% better model on training set

For real same case I would choose 4 or 5 dimention model (depending on the scale of dataset).
